## We plots figures for the  $R^n$ model for the relations to find $\beta_2$


### 0. Basis set up

```python
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt 
# style and notebook integration of the plots
import seaborn as sns
%matplotlib inline
plt.style.use(['science','latex-sans'])
print(plt.style.available)
```

```python
def f1(z,n):
    return (np.log10(np.abs((12**(1-n)*z**(2-2*n))/(n-2))))
def f2(z,n):
    return ((12**(1-n)*z**(2-2*n))/(n-2))
def f3(z,n):
    return np.abs((12**(1-n)*z**(2-2*n))/(n-2))
```

### 1. For the case $n<0$

```python
z_space = np.linspace(1, 65, 1000) # range = (0, 65)
fig, ax = plt.subplots()
ax.plot(z_space, f2(z_space, -1), 'b-', lw=0.55, label =r'$n = -1$')
ax.plot(z_space, f2(z_space, -2), 'r', linestyle='--',lw= 0.55, label =r'$n = -2$')
ax.plot(z_space, f2(z_space, -3), 'g-.', lw=0.55, label =r'$n = -3$')
ax.plot(z_space, f2(z_space, -4), 'purple', linestyle='dotted',lw=0.55, label =r'$n = -4$')
#ax.set_xscale('log')
ax.set_yscale('symlog')
ax.set_yticks([-1e23,-1e20,-1e15, -1e10, -1e5, \
               -1e2, -1e1])
ax.set_xlim(min(z_space), max(z_space))
ax.set_ylim(-1e23, -1e1)
ax.set_xlabel(r'$\zeta$',)
ax.set_ylabel(r'$\beta_2$')
#ax.set_yticklabels(y, fontsize=20)
#ax.set_xticklabels(ax, fontsize=6)
ax.tick_params(axis='x', labelsize=7)
ax.tick_params(axis='y', labelsize=7)
ax.legend(loc = 'upper right',fontsize = 5,frameon=True)
#plt.savefig('figspython/n-0-beta-zeta-1.pdf', bbox_inches = 'tight')
plt.show()
```

```python

z_space = np.linspace(1, 65, 1000) # range = (0, 65)
p1, = plt.plot(z_space, f2(z_space, -1/5), 'b-', lw=0.55, label =r'$n = -1/5$')
p2, = plt.plot(z_space, f2(z_space, -1/4), 'r', linestyle='--',lw= 0.55, label =r'$n = -1/4$')
p3, = plt.plot(z_space, f2(z_space, -1/3), 'g-.', lw=0.55, label =r'$n = -1/3$')
p4, = plt.plot(z_space, f2(z_space, -1/2), 'purple', linestyle='dotted',lw=0.55, label =r'$n =-1/2$')

plt.yscale('symlog')
plt.xlim(1, 65)
plt.xlabel(r'$\zeta$')
plt.ylabel(r'$\beta_2$')

plt.yticks(fontsize = 7)
plt.xticks(fontsize = 7)

plt.legend( loc = 'upper right',fontsize=5,frameon=True)
#plt.savefig('figspython/n-0-beta-zeta-2.pdf', bbox_inches = 'tight')
plt.show()
```
