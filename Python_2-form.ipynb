## We plot the $\Sigma/H$ from different models of 2-form anisotropic inflation


```python
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt 
# style and notebook integration of the plots



import seaborn as sns
%matplotlib inline
plt.style.use(['science','ieee'])
#print(plt.style.available)
```

```python
def fc(r, l):    
    return np.abs((-l**(2)-2*r*l+2)/(l**(2)+5*r*l+6*r**(2)+4))
def fDBI(g,r,l):
    return np.abs((2*g-l*(l+2*r))/(4*g+(l+2*r)*(l+3*r)))
def fk(r, l):
    return np.abs((-96*r**(2)-3*(-l-2*r)*(-3*l-3*r+np.sqrt(l**(2)-22*l*r+361*r**(2))))/(4*(3*l**(2)+9*l*r-42*r**(2))))
```

```python
l_space = np.linspace(0, 0.5, 1000) # range = (1e-2, 1e4)
plt.plot(l_space, fc(50, l_space), 'g-', label = r'Canonical')
plt.plot(l_space, fDBI(1.5,50, l_space), 'b--', label = r'DBI')
plt.plot(l_space, fk(50,l_space), 'r-.', label = r'$k$-inflation')

#plt.yscale('log')
plt.xlim(0.03, 0.5)
plt.xlabel(r'$|\lambda|$')
plt.ylabel(r'$|\Sigma/H|$')
plt.legend(loc = 'upper left')
plt.savefig('figspython/lambda-ani.pdf', bbox_inches = 'tight')

plt.show()
```

```python
r_space = np.linspace(0, 100, 1000)#range = (1e-2, 1e4)
plt.plot(r_space, fc(r_space, 0.1), 'g-', label = r'Canonical')
plt.plot(r_space, fDBI(1.5, r_space, 0.1), 'b--', label = r'DBI')
plt.plot(r_space, fk(r_space, 0.1),'r-.', label = r'$k$-inflation')

#plt.yscale('log')
plt.xlim(20,max(r_space))
plt.ylim(0,0.0009)
plt.xlabel(r'$\rho$')
plt.ylabel(r'$|\Sigma/H|$')
plt.legend(loc = 'upper right')
plt.savefig('figspython/rho-ani.pdf', bbox_inches = 'tight')
plt.show()
```

## 1.2 - Full Range

```python
def fc1(r, l):    
    return np.abs((2*(l**(2)+2*r*l-4))/(l**(2)+8*r*l+12*r**(2)+8))
def fc2(r, l):    
    return np.abs((-l**(2)-2*r*l+2)/(l**(2)+5*r*l+6*r**(2)+4))

def fk1(r,l):
    return np.abs((3*l**(2)+6*l*r+24*r**(2)+3*(-l-2*r)*np.sqrt(l**(2)-4*l*r+16*r**(2)))/(3*l**(2)+12*l*r-12*r**(2)))
                  
def fk2(r, l):
    return np.abs((-96*r**(2)-3*(-l-2*r)*(-3*l-3*r+np.sqrt(l**(2)-22*l*r+361*r**(2))))/(4*(3*l**(2)+9*l*r-42*r**(2))))

def fDBI1(g,r,l):
    return np.abs((2*(l**(2)+2*r*l-4*g))/(l**(2)+8*r*l+12*r**(2)+8*g))
def fDBI2(g,r,l):
    return np.abs((2*g-l*(l+2*r))/(4*g+(l+2*r)*(l+3*r)))
```

```python
categories = ['Canonical', r'$k$-inflation', 'DBI']
csfont = {'fontname':'Comic Sans MS'}
hfont = {'fontname':'Helvetica'}


l_space = np.linspace(0, 0.5, 1000) # range = (1e-2, 1e4)
p1, = plt.plot(l_space, fc1(50, l_space), 'g-',lw=0.65)
p2, = plt.plot(l_space, fDBI1(1.5,50, l_space), 'b', linestyle='-',lw=0.65)
p3, = plt.plot(l_space, fk1(50,l_space), 'r', linestyle='-',lw=0.65)
p4, = plt.plot(l_space, fc2(50, l_space), 'g', linestyle='--',lw=0.65)
p5, = plt.plot(l_space, fDBI2(1.5,50, l_space), 'b--',lw=0.65)
p6, = plt.plot(l_space, fk2(50,l_space), 'r--',lw=0.65)
p7, = plt.plot([0], marker='None',
           linestyle='None', label='dummy-tophead')
p8, = plt.plot([0],  marker='None',
           linestyle='None', label='dummy-empty')
#plt.yscale('log')
plt.xlim(0.06, 0.5)
plt.xlabel('$|\lambda|$')
plt.ylabel(r'$|\Sigma/H|$')

leg4 = plt.legend([p7, p1, p3, p2, p7, p4, p6, p5],
              [r'One-form'] + categories + [r'Two-form'] + categories,
              loc=2, ncol=2,fontsize="5") # Two columns, horizontal group labels

#plt.legend(loc = 'upper left')
plt.savefig('figspython/lambda-ani-12.pdf', bbox_inches = 'tight')
plt.show()
```
